```python
import matplotlib.pyplot as plt
```

```python
m = [10,20,30,40]
equi = [138,634, 1480, 2674]
non_equi = [132,612, 1447, 2634]
```

```python
plt.plot(m,equi,'.-', label = 'Equidistant')
plt.plot(m,non_equi,'r.--', label = 'Non-Equidistant')
plt.xticks = m
plt.ylabel('Iteration at convergence')
plt.xlabel('Mesh size $m x m$')
plt.legend()
plt.savefig('Iteration_comp.png')
```
